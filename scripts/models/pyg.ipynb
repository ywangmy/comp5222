```python
from superglue import *

batch_size = 4
num_nodes = 6
nfeatures = 50

edge_index, edge_type = generate_edges_union(num_nodes, num_nodes, True)

x = torch.randn(batch_size, num_nodes, nfeatures)

batch_list = []
from torch_geometric.data import Batch, Data

for i in range(batch_size):
    batch_list.append(Data(x=x[i, :, :], edge_index=edge_index, edge_type=edge_type))
batch = Batch.from_data_list(batch_list)

print(batch.edge_index.shape)

k = 0.5
batch_list = Batch.to_data_list(batch)
attention_weights = torch.randn(batch.edge_index.shape[1])
id_start = 0
for b in range(batch_size):
    num_edges = batch_list[b].edge_index.shape[-1]
    id_end = id_start + num_edges
    attention_weights_now = attention_weights[id_start:id_end]
    print(attention_weights_now.shape)
    topk_ids = torch.topk(attention_weights_now, int(k * num_edges)).indices.cuda()
    now_edge_index = batch_list[b].edge_index
    print(now_edge_index.shape)
    print(topk_ids)
    print(now_edge_index[:, topk_ids])
```
